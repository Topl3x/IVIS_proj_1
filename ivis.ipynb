```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv('data.csv')

# Extract the columns to be plotted
A = df['Alias']
B = df['Interest']
C = df['Information visualization']
D = df['Statistical']
E = df['Mathematics']
F = df['Drawing and artistic']
G = df['Computer usage']
H = df['Programming']
I = df['Computer graphics programming']
J = df['Human-computer interaction programming']
K = df['User experience evaluation']
L = df['Communication']
M = df['Collaboration']
N = df['Code repository']

TOT = []

for i in range(len(A)):
    TOT.append(C[i] + D[i] + E[i] + F[i] + G[i] + H[i] + I[i] + J[i] + K[i] + L[i] + M[i] + N[i])

def get_color(interest):
    if interest == 'sports':
        return 'blue'
    if interest == 'games':
        return 'orange'
    if interest == 'arts':
        return 'red'
    if interest == 'movies':
        return 'green'
    if interest == 'tech':
        return 'black'
    if interest == 'none':
        return 'gray'
    return None

class Person:
  def __init__(self, alias, interest, score):
    self.alias = alias
    self.interest = interest
    self.score = score
    
persons = []
for i in range(len(A)):
    persons.append(Person(A[i], B[i], TOT[i]))  

bar_colors = []
for person in persons:
    bar_colors.append(get_color(person.interest))

# Plot the bar graph
plt.ylim([0, 100])
plt.xticks(rotation=90)
plt.bar(A, TOT, color=bar_colors)

# Add labels and title to the graph
plt.xlabel('Alias');
plt.ylabel('Total score');
plt.title('Members skillsets');

# # Show the plot
# plt.show()
```

```python
def sort_by_score(person):
    return person.score
    
sorted_persons = sorted(persons, key=sort_by_score)

A_sorted = []
TOT_sorted = []

for i in range(len(A)):
    A_sorted.append(sorted_persons[i].alias)
    TOT_sorted.append(sorted_persons[i].score)
    
bar_colors = []
for person in sorted_persons:
    bar_colors.append(get_color(person.interest))

plt.ylim([0, 100])
plt.xticks(rotation=90)
plt.bar(A_sorted, TOT_sorted, color=bar_colors);
plt.xlabel('Alias');
plt.ylabel('Total score');
plt.title('Members skillsets');
```

```python
my_labels = ["sports","games","arts","movies","tech","none"]
XX = []
for i in my_labels:
    for p in sorted_persons:
        if p.interest == i:
            XX.append(p)

s = []
t = []
u = []

for p in XX:
    s.append(p.alias)
    t.append(p.score)
    u.append(get_color(p.interest))

plt.ylim([0, 100])
plt.xticks(rotation=90)
plt.bar(s, t, color=u)
plt.xlabel('Alias');
plt.ylabel('Total score');
plt.title('Members skillsets');
```

```python
SELECTED_GROUP = [0,13,17,23,40]
interests_distribution = { "sports" : 0, "games" : 0, "arts" : 0, "movies" : 0, "tech" : 0, "none" : 0 }


for x in SELECTED_GROUP:
    interests_distribution[persons[x].interest] += 20
    
p = []
for key in interests_distribution:
    p.append(interests_distribution[key])

plt.pie(p, startangle = 90, colors=['blue', 'orange', 'red', 'green', 'black', 'gray']);
plt.legend(my_labels, loc="best");
```
